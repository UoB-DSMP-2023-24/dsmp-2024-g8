## 序列相似性聚类
这种方法基于CDR3序列之间的相似性进行聚类。可以使用局部或全局序列比对算法，如BLAST或Smith-Waterman算法，计算序列间的相似性得分，然后使用这些得分进行聚类，常见的聚类算法包括层次聚类（Hierarchical clustering）和K-均值聚类（K-means clustering）。

## Levenshtein

序列相似性计算：使用Levenshtein距离（一种编辑距离）来衡量CDR3序列之间的相似性。Levenshtein距离衡量将一个字符串转换成另一个字符串所需的最少单字符编辑（插入、删除或替换）。
聚类：基于计算得到的距离矩阵，我们可以使用层次聚类方法来对序列进行聚类。层次聚类是一种常用的聚类技术，可以根据相似性度量聚集对象。

```python
# Select a subset of CDR3 sequences for demonstration purposes
cdr3_subset = vdjh_df['cdr3'].iloc[:50].tolist()
```

```python
from difflib import SequenceMatcher

def sequence_similarity(a, b):
    """Compute the similarity score between two sequences using SequenceMatcher."""
    return SequenceMatcher(None, a, b).ratio()

# Calculate the similarity matrix using SequenceMatcher
similarity_matrix = np.zeros((len(cdr3_subset), len(cdr3_subset)))

for i in range(len(cdr3_subset)):
    for j in range(i, len(cdr3_subset)):  # The matrix is symmetric
        similarity = sequence_similarity(cdr3_subset[i], cdr3_subset[j])
        # Convert similarity to distance
        distance = 1 - similarity
        similarity_matrix[i, j] = distance
        similarity_matrix[j, i] = distance

# Show a portion of the similarity (converted to distance) matrix
similarity_matrix[:5, :5]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist, squareform

# Perform hierarchical clustering using the updated distance matrix based on SequenceMatcher
Z_updated = linkage(squareform(similarity_matrix), method='complete', metric='precomputed')

# Decide on a cutoff distance to define clusters
cutoff_distance_updated = 0.5  # This value can be adjusted based on the dendrogram or specific requirements

# Form flat clusters from the hierarchical clustering defined by the given linkage matrix
clusters_updated = fcluster(Z_updated, t=cutoff_distance_updated, criterion='distance')

# Show the first few cluster assignments based on the updated distance matrix
clusters_updated[:10]
```

```python
from umap import UMAP

# Initialize UMAP. We use the precomputed metric since we are providing a distance matrix
umap_model = UMAP(n_components=2, metric='precomputed', random_state=42)

# Fit UMAP on the computed distance matrix (note: we are using the similarity matrix as a distance matrix here)
umap_embedding = umap_model.fit_transform(similarity_matrix)

# Plot the UMAP embedding with cluster labels
plt.figure(figsize=(8, 6))
plt.scatter(umap_embedding[:, 0], umap_embedding[:, 1], c=clusters_updated, cmap='Spectral', s=50)
plt.title('UMAP Projection of CDR3 Clusters')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

## 基于距离的聚类
这种方法先定义一个序列距离度量（如编辑距离或海明距离），然后基于这些距离使用聚类算法

k-mer 频率和 DBSCAN 都是依据数据点之间的距离进行聚类。

## k-mer

数据降维：如果序列长度不一，可以将所有序列编码到固定长度的向量中，可能通过诸如k-mer频率等方法。k-mer频率:序列中k长度子序列的出现频率

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import numpy as np

# Function to generate k-mers for a sequence
def generate_kmers(sequence, k=5):
    return [sequence[i:i+k] for i in range(len(sequence) - k + 1)]

# Generating k-mers for each sequence
k = 5
vdj_df['kmer'] = vdj_df['cdr3'].apply(lambda x: generate_kmers(x, k))

# Flattening the list of k-mers for CountVectorizer
kmer_list = [' '.join(kmers) for kmers in vdj_df['kmer']]

# Vectorizing the k-mers to create a feature matrix
vectorizer = CountVectorizer(analyzer='word', token_pattern=r"(?u)\b\w+\b")
X = vectorizer.fit_transform(kmer_list)

# Clustering (using KMeans as an example, number of clusters arbitrarily set to 5)
# Adjust n_clusters based on expected number of clusters or use methods like Elbow to find an optimal number
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)

# Adding cluster labels to the dataframe
vdj_df['cluster'] = kmeans.labels_

# Checking the distribution of sequences across the clusters
cluster_distribution = vdj_df['cluster'].value_counts()

cluster_distribution
```

评估 k-mer 长度:k 的最佳值对于确保聚类结果的有效性和解释性非常重要。一个常见的方法是轮廓系数。
轮廓系数 (Silhouette Score)：衡量每个点的聚类紧密度和分离度，值范围为 [-1, 1]，值越大表示聚类效果越好。

```python
from sklearn.metrics import silhouette_score

# Function to perform clustering and calculate silhouette score for different k-mer lengths
def calculate_silhouette_scores(data, k_values):
    scores = {}
    for k in k_values:
        # Generate k-mers
        vdj_df['kmer'] = vdj_df['cdr3'].apply(lambda x: generate_kmers(x, k))
        kmer_list = [' '.join(kmers) for kmers in vdj_df['kmer']]
        
        # Vectorize k-mers
        vectorizer = CountVectorizer(analyzer='word', token_pattern=r"(?u)\b\w+\b")
        X = vectorizer.fit_transform(kmer_list)
        
        # Cluster (using a reasonable default of n_clusters to simplify comparison)
        n_clusters = 5
        kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
        
        # Calculate silhouette score
        score = silhouette_score(X, kmeans.labels_)
        scores[k] = score
    
    return scores

# Range of k values to try
k_values = range(3, 11)

# Calculate silhouette scores for each k
silhouette_scores = calculate_silhouette_scores(vdj_df, k_values)

silhouette_scores
```

```python
# Dimensionality reduction with UMAP
umap = UMAP(n_neighbors=5, min_dist=0.3, metric='correlation')
X_reduced = umap.fit_transform(X.toarray())

# Plotting the UMAP reduced dimensions colored by cluster label
plt.figure(figsize=(10, 8))
for i in range(n_clusters):
    plt.scatter(X_reduced[vdj_df['cluster'] == i, 0], X_reduced[vdj_df['cluster'] == i, 1], label=f'Cluster {i+1}', alpha=0.7)
plt.title('CDR3 Sequences Clustering Visualization with UMAP')
plt.legend()
plt.show()
```

```python
silhouette_coef = silhouette_score(X, vdj_df['cluster'])

print(silhouette_coef)
```

聚类结果整体上比较弱，聚类内的凝聚度和聚类间的分离度都不是很理想。这个值接近于 0，可能表明许多样本在它们自己的聚类中并不是很紧密，或者与其他聚类的距离不够远。

## DBSCAN

```python
from sklearn.cluster import DBSCAN

# 使用 DBSCAN 聚类
# 这里的 eps 和 min_samples 是示例值，实际应用中需要根据数据进行调整
dbscan = DBSCAN(eps=0.5, min_samples=5)

# 使用 k-mer 频率特征进行聚类
dbscan_clusters = dbscan.fit_predict(X)

# 将聚类结果添加到数据帧中
vdj_df['dbscan_cluster'] = dbscan_clusters

# 查看聚类分布
cluster_distribution = pd.value_counts(vdj_df['dbscan_cluster'])

print(cluster_distribution)
```

在 DBSCAN 聚类中，-1 表示噪声点，即这些点在给定的 eps（领域大小）和 min_samples（形成稠密区域所需的最小样本数）参数下没有被分配到任何具体的聚类中。在聚类结果中，有 4412 个点被标记为噪声，这表明可能需要调整 DBSCAN 的参数来减少噪声点的数量，以便更好地捕捉数据中的聚类结构。
