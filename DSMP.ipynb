```python
import pandas as pd

# Example dataframe
data = {
    'gene': ['TRA', 'TRB', 'TRA', 'TRB', 'TRA', 'TRB', 'TRA', 'TRB'],
    'cdr3': ['CAVPSGAGSYQLTF', 'CASSFEPGQGFYSNQPQHF', 'CAVPSGAGSYQLTF', 'CASSFEPGQGFYSNQPQHF', 'CAVPSGAGSYQLTF', 'CASSFEPGQGFYSNQPQHF', 'CAVPSGAGSYQLTF', 'CASSFEPGQGFYSNQPQHF'],
    'v.segm': ['TRAV20*01', 'TRBV13*01', 'TRAV20*01', 'TRBV13*01', 'TRAV20*01', 'TRBV13*01', 'TRAV20*01', 'TRBV13*01'],
    'j.segm': ['TRAJ28*01', 'TRBJ1-5*01', 'TRAJ28*01', 'TRBJ1-5*01', 'TRAJ28*01', 'TRBJ1-5*01', 'TRAJ28*01', 'TRBJ1-5*01'],
    'species': ['HomoSapiens', 'HomoSapiens', 'HomoSapiens', 'HomoSapiens', 'HomoSapiens', 'HomoSapiens', 'HomoSapiens', 'HomoSapiens'],
    'subject_id': [200, 200, 200, 200, 200, 200, 200, 200]
}
df = pd.DataFrame(data)

# Group by 'subject_id'
grouped_df = df.groupby('subject_id')

# Create a new dataframe
new_df = pd.DataFrame(columns=['cdr3_TRA', 'v.segm_TRA', 'j.segm_TRA', 'cdr3_TRB', 'v.segm_TRB', 'j.segm_TRB', 'species', 'subject_id'])
for _, group in grouped_df:
    # Extract 'TRA' and 'TRB' rows
    tra_row = group[group['gene'] == 'TRA'].squeeze()
    trb_row = group[group['gene'] == 'TRB'].squeeze()
    
    # Combine rows
    combined_row = {
        'cdr3_TRA': tra_row['cdr3'], 'v.segm_TRA': tra_row['v.segm'], 'j.segm_TRA': tra_row['j.segm'],
        'cdr3_TRB': trb_row['cdr3'], 'v.segm_TRB': trb_row['v.segm'], 'j.segm_TRB': trb_row['j.segm'],
        'species': tra_row['species'], 'subject_id': tra_row['subject_id']
    }
    
    # Append combined row to new dataframe
    new_df = new_df.append(combined_row, ignore_index=True)

new_df
```
